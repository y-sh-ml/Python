```python
import pytz
from datetime import datetime

def write_log(n, m):
    tz = pytz.timezone('Asia/Tokyo')
    current_time = datetime.now(tz).strftime('%Y-%m-%d %H:%M:%S')
    with open('log_file/log.txt', 'a') as log_file:
        log_file.write(f'batch {n}-{m}: completed {current_time}\n')
```
