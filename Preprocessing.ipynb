```python
# ã‚«ãƒ©ãƒ å†…ã®æ–‡å­—ã‚’æ•°å­—ã«å¤‰æ›ã™ã‚‹

def code_transform(x):
    if x == 'F':
        y = 0
    elif x == 'M':
        y = 1
    else:
        y = 2
        
    return y

df['CODE_GENDER_INT'] = df.CODE_GENDER.apply(lambda x: code_transform(x))
df['CODE_GENDER_INT'].head()
```
